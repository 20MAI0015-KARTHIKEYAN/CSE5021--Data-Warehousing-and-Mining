```python
#importing the Required packages
from bs4 import BeautifulSoup as bs
import requests as rq
import pandas as pd
```

```python
#imporing the contents of the page
requesting_page = rq.get("https://en.wikipedia.org/wiki/COVID-19_pandemic_in_Tamil_Nadu")
```

```python
#Extracting the contents to a variale
page_text = requesting_page.content
```

```python
#Using Beautiful Soup parsing the HTML
soup = bs(page_text, 'html.parser')
```

EXTRACTING CONTANTS OF TABLE 1

```python
#Tables were nested here in the webpage so we need to find the right ones as per the requirement and we have to hardcode it as we are not using any api to scrape contents of the page
tnbd = soup.find("div", attrs={"id":"covid19-container"}).find("table")
```

```python
#finding all the tr tags
trtag = tnbd.find_all("tr")[3:40]
```

```python
#Visualise the data
print(trtag)
```

```python
#Assign lists for collecting data from a for loop
dt = []
dc = []
dth = []
rcv = []
act = []
pop = []
m = []
```

```python
#Using a for loop to scrape the data
for i in trtag:
  dat = i.find_all("td")
  dt.append(dat[0].find("a").text.strip())
  dc.append(dat[1].text.strip())
  dth.append(dat[2].text.strip())
  rcv.append(dat[3].text.strip())
  act.append(dat[4].text.strip())
  pop.append(dat[5].text.strip())
  m.append(dat[6].text.strip())
```

```python
#Extracted data from table 
for i in range (0,len(dt)):
  print(dt[i]+"-"+dc[i]+"-"+dth[i]+"-"+rcv[i]+"-"+act[i]+"-"+pop[i]+"-"+m[i])
```

```python
#dictionary creation for putting data in the form of a dataframe
tn_district_data={
    'DISTRICT':dt,
    'DIAGNOSED_CASES':dc,
    'DEATHS':dth,
    'RECOVERED_CASES':rcv,
    'ACTIVE_CASES':act,
    'POPULATION':pop,
    'CASES_PER_M':m
}
```

```python
#Converting the data into dataframe
dt_data = pd.DataFrame(tn_district_data)
```

```python
#CSV format conversion
dt_data.to_csv("TN_DISTRICTWISE_DATA.csv")
```

EXTRACTING CONTANTS OF TABLE 2

```python
#Tables were nested here in the webpage so we need to find the right ones as per the requirement and we have to hardcode it as we are not using any api to scrape contents of the page
tn_stpw=soup.find('table',attrs={'class':'multicol','role':'presentation'}).find('table')
```

```python
#Extractinng the table body
tn_stpw=tn_stpw.find('tbody')
```

```python
#finding all the tr tags
trtag1 = tn_stpw.find_all('tr')[1:]
```

```python
#Visuaalizing the data
print(trtag1)
```

```python
#initializing lists to hold data while looping
wk = []
nt = []
pv = []
pp = []
ct = []
```

```python
#Using a for loop to scrape the data
for i in trtag1:
  dat1=i.find_all("td")
  wk.append(i.find("th").text.strip())
  nt.append(dat1[0].text.strip())
  pv.append(dat1[1].text.strip())
  pp.append(dat1[2].text.strip())
  ct.append(dat1[3].text.strip())
```

```python
#dictionary creation for putting data in the form of a dataframe
tn_sample_tested_per_week={
    'WEEK':wk,
    'NEW_TESTS':nt,
    'POSITIVE':pv,
    'POSITIVE_PERCENTAGE':pp,
    'CUMULATIVE_TESTS':ct
}
```

```python
#Converting the data into dataframe
tn_stpw_data = pd.DataFrame(tn_sample_tested_per_week)
```

```python
#Converting to CSV
tn_stpw_data.to_csv("SAMPLES_TESTED_PER_WEEK.csv")
```

```python
#Tables were nested here in the webpage so we need to find the right ones as per the requirement and we have to hardcode it as we are not using any api to scrape contents of the page
mep = soup.find("div",attrs={"class":"mw-parser-output"}).find_all("table")
```

```python
#Extracting the appropriate table
mep=mep[4]
```

```python
#Extractinng the table body
mep=mep.find("tbody")
```

```python
#finding all the tr tags
trtag2 = mep.find_all("tr")[1:]
```

```python
#Visualizing the trtag
print(trtag2)
```

```python
#initializing lists to hold data while looping
dt = []
ev = []
```

```python
#Using a for loop to scrape the data
for i in trtag2:
  dt.append(i.find("th").text.strip())
  ev.append(i.find("td").text.replace("<br/>","").strip())
```

```python
#dictionary creation for putting data in the form of a dataframe
tn_mep={
    'DATE':dt,
    'EVENT':ev
}
```

```python
#Converting the data into dataframe
tn_mep_data = pd.DataFrame(tn_mep)
```

```python
#converting the dataframe to csv
tn_mep_data.to_csv("MAJOR_EVENTS_TILL_APRIL_30.csv")
```
