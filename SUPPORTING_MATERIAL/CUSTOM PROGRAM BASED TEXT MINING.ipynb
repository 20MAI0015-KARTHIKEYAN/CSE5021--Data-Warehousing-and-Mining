MINING WITHOUT API

```python
import requests as r
from bs4 import BeautifulSoup
import pandas as pd
```

```python
wiki_page_request = r.get("https://en.wikipedia.org/wiki/COVID-19_pandemic_in_Tamil_Nadu")
wiki_page_text = wiki_page_request.content
soup = BeautifulSoup(wiki_page_text, 'html.parser')
```

### TABLE 1 : COVID-19 pandemic in Tamil Nadu by district

```python
Table=soup.find('div',attrs={'id':'covid19-container'}).find('table')
Trs=Table.find_all('tr')[3:40]
```

```python
Trs
```

```python
District=[]
Diagnosed_cases=[]
Deaths=[]
Recovered_cases=[]
Active_cases=[]
Population=[]
Cases_per_M=[]
```

```python
for row in Trs:
    Tds=row.find_all('td')
    District.append(Tds[0].find('a').text.strip())
    Diagnosed_cases.append(Tds[1].text.strip())
    Deaths.append(Tds[2].text.strip())
    Recovered_cases.append(Tds[3].text.strip())
    Active_cases.append(Tds[4].text.strip())
    Population.append(Tds[5].text.strip())
    Cases_per_M.append(Tds[6].text.strip())
```

```python
Covid_district_wise={
    'District':District,
    'Diagnosed_cases':Diagnosed_cases,
    'Deaths':Deaths,
    'Recovered_cases':Recovered_cases,
    'Active_cases':Active_cases,
    'Population':Population,
    'Cases_per_M':Cases_per_M
}

Covid_district_data=pd.DataFrame(Covid_district_wise)
Covid_district_data.to_csv('Covid_district_data.csv')
Covid_district_data.head()
```

### TABLE 2: Number of samples tested per week

```python
Table=soup.find('table',attrs={'class':'multicol','role':'presentation'}).find('table')
```

```python
Table=Table.find('tbody')
Trs=Table.find_all('tr')[1:]
```

```python
Trs
```

```python
Week=[]
New_tests=[]
Positive=[]
Positive_Percentage=[]
Cumulative_tests=[]
```

```python
for row in Trs:
    Week.append(row.find('th').text.strip())
    tds=row.find_all('td')
    New_tests.append(tds[0].text.strip())
    Positive.append(tds[1].text.strip())
    Positive_Percentage.append(tds[2].text.strip())
    Cumulative_tests.append(tds[3].text.strip())
```

```python
Samples_Tested_weekly={
    'Week':Week,
    'New_tests':New_tests,
    'Positive':Positive,
    'Positive_Percentage':Positive_Percentage,
    'Cumulative_tests':Cumulative_tests
}
Samples_Tested_weekly_data=pd.DataFrame(Samples_Tested_weekly)
Samples_Tested_weekly_data.to_csv('Samples_Tested_weekly_data.csv')
Samples_Tested_weekly_data.head()
```

### Table 3: Major events in COVID19 pandemic till 30th April

```python
Table=soup.find('div',attrs={'class':'mw-parser-output'}).find_all('table')
Table=Table[4]
Table=Table.find('tbody')
```

```python
Trs=Table.find_all('tr')[1:]
Trs
```

```python
Date=[]
Event=[]
```

```python
for row in Trs:
    Date.append(row.find('th').text.strip())
    Event.append(row.find('td').text.replace('<br/>','').strip())
```

```python
Events_Till_April={
    'Date':Date,
    'Event':Event
}
Events_Till_April_data=pd.DataFrame(Events_Till_April)
Events_Till_April_data.to_csv('Events_Till_April_data.csv')
Events_Till_April_data.head()
```
